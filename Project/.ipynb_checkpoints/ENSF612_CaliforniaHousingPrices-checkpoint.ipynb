```python
#Import libaries
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
#Convert CSV to pandas dataframe
df = pd.read_csv("housing.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
#Data Preprocessing
#From the info above we can see that total_bedroom has null values. We can try to fill it with median or mean values
# but we first should determine which approach is more suitable. We can see that the maximum amount of total_bedroom is
# 6445 while the minimum amount is 537 indicating the data might be skewed.  
df['total_bedrooms'] = df['total_bedrooms'].fillna((df['total_bedrooms'].median()))
# df
```

```python
#We can create a heatmap to determine which features have the biggest impact on on our target variable of median_house_value
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title('Correlation Heatmap of California Housing Data Set')
plt.show()

#From the heatmap, we can see that median income has the highest importance to median_house_value
```

```python
df.hist(bins=30,figsize=(15,15))
```

```python
#Machine learning algorithms can't handle categorical variables unless we convert
#them to numerical values
from sklearn.preprocessing import OneHotEncoder
#creating instance of one-hot-encoder
encoder = OneHotEncoder()
#perform one-hot encoding on columns 
encoder_df = pd.DataFrame(encoder.fit_transform(df[['ocean_proximity']]).toarray())
#merge one-hot encoded columns back with original DataFrame
df = df.join(encoder_df)
#drop columns with strings
df.drop('ocean_proximity', axis=1, inplace=True)
#showing new head
df.head()
```

```python
#From the histograms we can see that many attributes are tail heavy meaning that they tend to lean to one side of the axis 
#Such as median_income has a positive right trail, so we can apply a log transformation on it in hopes that it will normalize
# the data
df['median_income'] = np.log(df['median_income'])
df['median_house_value'] = np.log(df['median_house_value'])
```

```python
#Since we want to predict the housing prices that will be our depedent variable/target variable
#The X attributes (independent features) would include the various characteristics of each niehgbourhood
#that we will be using to predict the housing prices. 
x = df.drop('median_house_value', axis = 1)
x.columns = x.columns.astype(str)
y = df['median_house_value']
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x_scaled = pd.DataFrame(scaler.transform(x), index= x.index, columns= x.columns)
```

# Splitting the data

```python

X_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 0)
## Check Shapes of these Sets
print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
print('X_test shape:', x_test.shape)
print('y_test shape:', y_test.shape)
```

# Using Linear Regression as Machine Learning model with Feature Scaling

Linear regresssion is the first type of machine learning algorithm used in forecast continous numerical outcome variable based on one or more predictor variables. This is achieved by minimizing the sum of the squared deviations between the predicted and actual target values. 


```python
from sklearn.linear_model import LinearRegression #importing linear regression
lr = LinearRegression() # Initializing the model
model = lr.fit(X_train, y_train) # Fitting the model to the training data
```

Calculating the RMSE and R2 score in the regression model

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred_lr = lr.predict(x_test)
mse = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse)
print('RMSE:', rmse_lr)
```

```python
r2_lr = r2_score(y_test, y_pred_lr)
print('R-square:', r2_lr)
```

```python
# Print the coefficients and their corresponding feature names
for coef, feature_name in zip(model.coef_, x_scaled.columns):
    print(f'{feature_name}: {coef:.2f}')
```

# Using KNN  as Machine Learning model


K-Nearest Neighbors (KNN) is a machine learning algorithm used for classification and regression tasks. It operates by selecting the k closest instances to a new data point from the training set, and then taking the most common (in classification) or average (in regression) value among those k instances as the prediction for the new data point.




```python
from sklearn.neighbors import KNeighborsRegressor

#creating a KNN regressor
knn = KNeighborsRegressor()

# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
knn.fit(X_train, y_train)

# performing predictions on the test dataset
y_pred_knn = knn.predict(x_test)

#RMSE and R-square metrics
rmse_knn = np.sqrt(mean_squared_error(y_test, y_pred_knn))
r2_knn = r2_score(y_test, y_pred_knn)

print('RMSE:', rmse_knn)
print('R-square:', r2_knn)
```

# Using Random Forest as Machine Learning Model 

The Random Forest Regressor is an algorithm used for ensemble learning, which combines multiple decision trees to forecast continuous numerical target variables. Its operation entails constructing numerous decision trees, each on a distinct random subset of the training data. The algorithm then aggregates the predictions from all the trees to generate a final prediction.





```python
from sklearn.ensemble import RandomForestRegressor

# creating a RF Regressor
rf = RandomForestRegressor(n_estimators = 100)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
rf.fit(X_train, y_train)
  
# performing predictions on the test dataset
y_pred_rf = rf.predict(x_test)
  
# metrics are used to find accuracy or error
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print('RMSE:', rmse_rf)
print('R-square:', r2_rf)
```

```python
ml_models = ['log_reg', 'knn','rf']
rmse_values = [rmse_lr,rmse_knn,rmse_rf]
r2_score_values = [r2_lr, r2_knn, r2_rf]

scores = {'Models': ml_models, 'RMSE': rmse_values, 'R2 Score': r2_score_values}
score_values = pd.DataFrame(scores)
display(score_values)
```

From the three machine learning models, we are comparing the RMSE and R-Square values. From the models we can see that random forest regressor has the best RMSE and R2 score.
